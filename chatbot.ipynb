```python
def menu_chatbot():
    print("ðŸ¤– Welcome to Menu-Based Chatbot")

    while True:
        print("\n----- MENU -----")
        print("1. About Us")
        print("2. Services")
        print("3. Contact")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            print("We are a software company providing IT solutions.")

        elif choice == "2":
            print("Our Services:")
            print("- Web Development")
            print("- AI & ML Solutions")
            print("- Database Management")

        elif choice == "3":
            print("Contact us at: support@example.com")

        elif choice == "4":
            print("Thank you! Chatbot closed.")
            break

        else:
            print("Invalid choice. Please try again.")
```
